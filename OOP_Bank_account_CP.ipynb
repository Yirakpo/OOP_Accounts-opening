```python
print("Welcome to the Livinginspired Smart Banking System")
class Account:
    def __init__(self, account_balance=0):
        self.pin = input("Set your 4-digit PIN: ")
        if len(self.pin) != 4 or not self.pin.isdigit():
            raise ValueError('PIN must be a 4-digit number')
        print('PIN set successfully')

        self.account_number = input("Enter the account number: ")
        if len(self.account_number) != 10 or not self.account_number.isdigit():
             raise ValueError("Account number must be a number")

        self.account_holder = input("Enter the account holder's name: ")
        self.account_type = input("Enter the account type (savings or current): ")
        if self.account_type in ['savings', 'current']:
            print('Account created successfully')
        else:
            raise ValueError('Invalid account type! please enter a valid account type')
        self.account_balance = account_balance  # Initialize account balance

    def deposit(self, amount: float):
        if amount <= 0:
            print("Invalid deposit amount")
        else:
            self.account_balance += amount
            print(f"Deposited  {amount}. \nyour new balance is : ${self.account_balance}")

    def withdraw(self, amount: float):
        if self.account_balance >= amount:
            self.account_balance -= amount
            print(f"Withdrew  {amount}. \nYour new balance is : ${self.account_balance}")
        else:
            print(f"Insufficient funds to withdraw {amount}. \nYour current balance is : ${self.account_balance}")

    def check_balance(self):
        return self.account_balance

# Testing the Account class
if __name__ == "__main__":
    # Create an instance of the Account class with user inputs
    my_account = Account()

    # Perform operations using user input
    while True:
        print("\nChoose an operation:")
        print("1. Deposit ")
        print("2. Withdraw ")
        print("3. Check balance ")
        print("4. Exit")
        choice = input("Enter the number of the operation:  ")

        if choice == "1":
            amount = float(input("Enter the amount to deposit: "))
            my_account.deposit(amount)
        elif choice == "2":
            amount = float(input("Enter the amount to withdraw: "))
            my_account.withdraw(amount)
        elif choice == "3":
            balance = my_account.check_balance()
            print(f" your Current balance is: ${balance}")
        elif choice == "4":
            print("Thank you for banking with us!\nPlease send us your feedback\nOur mobile banking system is working 24/7")
            break
        else:
            print("Invalid choice. Please try again.")
```
